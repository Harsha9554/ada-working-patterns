```python
def get_project_id(url):
    try:

      slash_split = url.split('/')
      project_id = slash_split[4]
      task_split = slash_split[-1].split('?')
      task_id = task_split[0]
      args_and_qs_split = task_split[-1].split('&')
      q_args = args_and_qs_split[-1]
      assignment_id = args_and_qs_split[0].split('=')[-1]
      return project_id
    except:
      return None

def get_task_id(url):
    if not "https://worker.mturk.com/projects/" in url:
      return None
    try:
      slash_split = url.split('/')
      project_id = slash_split[4]
      task_split = slash_split[-1].split('?')
      task_id = task_split[0]
      args_and_qs_split = task_split[-1].split('&')
      q_args = args_and_qs_split[-1]
      assignment_id = args_and_qs_split[0].split('=')[-1]
      return task_id
    except:
      return None

def get_assignment_id(url):
    if not "https://worker.mturk.com/projects/" in url:
      return None
    try:
      slash_split = url.split('/')
      project_id = slash_split[4]
      task_split = slash_split[-1].split('?')
      task_id = task_split[0]
      args_and_qs_split = task_split[-1].split('&')
      q_args = args_and_qs_split[-1]
      assignment_id = args_and_qs_split[0].split('=')[-1]
      return assignment_id
    except:
      return None

def get_query_params(url):
    if not "https://worker.mturk.com/projects/" in url:
      return None
    try:
      slash_split = url.split('/')
      project_id = slash_split[4]
      task_split = slash_split[-1].split('?')
      task_id = task_split[0]
      args_and_qs_split = task_split[-1].split('&')
      q_args = args_and_qs_split[-1]
      assignment_id = args_and_qs_split[0].split('=')[-1]
      return q_args
    except:
      return None

mturk_pdf['project_id'] = mturk_pdf.URL.map(get_project_id)
mturk_pdf['task_id'] = mturk_pdf.URL.map(get_task_id)
mturk_pdf['assignment_id'] = mturk_pdf.URL.map(get_assignment_id)
mturk_pdf['query_params'] = mturk_pdf.URL.map(get_query_params)
mturk_pdf[(mturk_pdf['platform'] == 'MTURK')][["URL","project_id", "task_id", "assignment_id", "user", "time"]].sample(50)
```

```python
def get_time(task_data, subtype):
  """
  Safely retrieves the time for a given task subtype, considering empty DataFrames and NA values.
  """
  if task_data.empty:
    return pd.NA  # Return NA if the DataFrame is empty

  try:
    time_value = task_data[task_data["task_subtype"] == subtype]["time"].iloc[0]
    # Check for NA using pd.isna()
    if pd.isna(time_value):
      return pd.NA  # Return NA if the retrieved time is NA
    else:
      return time_value
  except IndexError:
    return pd.NA  # Return NA for missing events
```

```python

results_df = pd.DataFrame(columns=["task_id", "user", "started_at", "submitted_at", "time_taken"])
for task_id in user_task_ids:
    if task_id != 'tasks':
      task_data = mturk_pdf[mturk_pdf["task_id"] == task_id]

      started_at = get_time(task_data, "TASK_STARTED")
      submitted_at = get_time(task_data, "TASK_SUBMITED")

      try:
        time_taken = (submitted_at - started_at).total_seconds()
      except:
        time_taken = -1

      new_row_df = pd.DataFrame({
      "task_id": task_id,
      "user": target_user_id,
      "started_at": started_at,
      "submitted_at": submitted_at,
      "time_taken": time_taken
      }, index=[0])  # Create a DataFrame with a single row


      results_df = pd.concat([results_df, new_row_df], ignore_index=True)
results_df.fillna(value={"started_at": pd.NA, "submitted_at": pd.NA}, inplace=True)
```
